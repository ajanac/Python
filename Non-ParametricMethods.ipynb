```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline 
rcParams['figure.figsize']=4,3
plt.style.use('seaborn-whitegrid')
```

```python
cars=pd.read_csv('mtcars.csv')
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
# The variables cyl,vs,am,gear are ordinal variables
# Because there are numerical variables and has only set of possible values
# These variables looks like exhibiting non linear relation
X=cars[['cyl','vs','am','gear']]
sb.pairplot(X)
```

```python
cyl=cars.cyl
vs=cars.vs
am=cars.am
gear=cars.gear
# getting Spearman coefficient and p value
spearmanr_coefficient,p_value=spearmanr(cyl,vs)
# printing the coefficient value in 3 deciam point
print "Spearman Rank Correlation %0.3f" % (spearmanr_coefficient)
```

```python
spearmanr_coefficient,p_value=spearmanr(cyl,am)
# printing the coefficient value in 3 deciam point
print "Spearman Rank Correlation %0.3f" % (spearmanr_coefficient)
```

```python
spearmanr_coefficient,p_value=spearmanr(cyl,gear)
# printing the coefficient value in 3 deciam point
print "Spearman Rank Correlation %0.3f" % (spearmanr_coefficient)
```

<h3> Chi-square test for independence </h3>

```python
table=pd.crosstab(cyl,am)

from scipy.stats import chi2_contingency
chi2,p,dof,expected=chi2_contingency(table.values)
print 'Chi-square Statistics %0.3f p_value %0.3f' % (chi2,p)
```

```python
table=pd.crosstab(cyl,vs)

from scipy.stats import chi2_contingency
chi2,p,dof,expected=chi2_contingency(table.values)
print 'Chi-square Statistics %0.3f p_value %0.3f' % (chi2,p)
```

```python
table=pd.crosstab(cyl,gear)

from scipy.stats import chi2_contingency
chi2,p,dof,expected=chi2_contingency(table.values)
print 'Chi-square Statistics %0.3f p_value %0.3f' % (chi2,p)
```
