```python
# Packages for data analysis
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

# Packages for data visualization
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
# Tell the Jupyter notebook to use the GUI inside the notebook
# rather than using an external GUI
%matplotlib inline
# setting the global parameters
rcParams['figure.figsize']=8,4
sb.set_style('whitegrid')
```

<h3> Labeling Plot Elements</h3>

<h5> The functional Methods </h5>

```python
x=range(1,10)
y=[1,2,3,4,1.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
z=[1,2,3,4,0.5]
veh_type=['motorbike','bycycle','car','van','storller']
plt.pie(z,labels=veh_type)
plt.show()
```

<h5> The Object Oriented Method <h5>

```python
car_data=pd.read_csv('mtcars.csv')
cars_columns=['car_names','mpg','cyl','wt']
mpg=car_data.mpg
# creating a blank figure object
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(36))
ax.set_xticklabels(car_data.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallen per Car in mtcars data set')
ax.set_xlabel('car names')
ax.set_ylabel('Mileage')
```

<h3>Adding a legend to your plot</h3>

<h5> The Functional Method </h5>

```python
plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

<h5> Object Oriented Method </h5>

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(36))
ax.set_xticklabels(car_data.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallen per Car in mtcars data set')
ax.set_xlabel('car names')
ax.set_ylabel('Mileage')
ax.legend(loc='best')
```
