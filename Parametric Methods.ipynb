```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

# Pearson Coefficient
import scipy 
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize']=8,4
plt.style.use('seaborn-whitegrid')
```

<h3> The Pearson Correlation</h3>

```python
cars=pd.read_csv('mtcars.csv')
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
# Scatter Plot Matrix
sb.pairplot(cars)
```

```python
X=cars[['mpg','hp','qsec','wt']]
sb.pairplot(X)
```

<h3> Using scipy to calculate the Pearson correlation coefficient</h3>

```python
mpg=cars['mpg']
hp=cars['hp']
qsec=cars['qsec']
wt=cars['wt']

# Pearson Correlation for mpg and hp correlation pair
pearsonr_coefficient,p_value=pearsonr(mpg,hp)
print 'PearsonR Correlation Coefficient %0.3f' %(pearsonr_coefficient)
# Negative values menans variables are negatively correlated
```

```python
pearsonr_coefficient,p_value=pearsonr(mpg,qsec)
print 'PearsonR Correlation Coefficient %0.3f' % (pearsonr_coefficient)
# Linear correlation between mpg and qsec
```

```python
pearsonr_coefficient,p_value=pearsonr(mpg,wt)
print 'PearsonR Correlation Coefficient %0.3f' % (pearsonr_coefficient)
```
